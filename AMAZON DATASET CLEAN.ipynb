```python
import pandas as pd

# Load the dataset
file_path = "C:\\Users\\HP\\OneDrive\\Desktop\\cleant1\\amazon_products.csv"  # Update if needed
df = pd.read_csv(file_path)

# Function to clean numeric columns safely
def clean_numeric(column):
    # Convert to string only if not already numeric
    if column.dtype == "object":
        return pd.to_numeric(column.str.replace(r"[₹,]", "", regex=True), errors="coerce")
    return column  # If already numeric, return as is

# Apply only if the column exists
if "Price" in df.columns:
    df["Price"] = clean_numeric(df["Price"])

if "Reviews" in df.columns:
    df["Reviews"] = clean_numeric(df["Reviews"])

# Print data types to verify
print(df.dtypes)
```

```python
if "Discount" in df.columns:
    df["Discount"] = df["Discount"].str.extract(r"(\d+)").astype(float)

# Print the first few rows to check the output
print(df[["Discount"]].head())  
```

```python
import pandas as pd
import re

# Remove text after "Details"
if "Delivery Date" in df.columns:
    df["Delivery Date"] = df["Delivery Date"].str.split("Details").str[0].str.strip()

# Function to extract only the date
def extract_date(text):
    if pd.isna(text) or text.strip() == "":
        return "Unknown"
    
    match = re.search(r"\b(?:\d{1,2} [A-Za-z]+|\d{1,2} [A-Za-z]+ - \d{1,2} [A-Za-z]+)\b", str(text))
    return match.group(0) if match else "Unknown"

# Apply function
if "Delivery Date" in df.columns:
    df["Delivery Date"] = df["Delivery Date"].apply(extract_date)

# Print the cleaned "Delivery Date" column
print(df[["Delivery Date"]].head())  
```

```python
import pandas as pd

# Fill missing values
if "Rating" in df.columns:
    df["Rating"] = df["Rating"].fillna(df["Rating"].median())

if "Reviews" in df.columns:
    df["Reviews"] = df["Reviews"].fillna(0)

if "Discount" in df.columns:
    df["Discount"] = df["Discount"].fillna(0)

if "Availability" in df.columns:
    df["Availability"] = df["Availability"].fillna("Unknown")

# Print the updated columns
print("Updated Rating Column:")
print(df[["Rating"]].head())  # Print first few values

print("\nUpdated Reviews Column:")
print(df[["Reviews"]].head())

print("\nUpdated Discount Column:")
print(df[["Discount"]].head())

print("\nUpdated Availability Column:")
print(df[["Availability"]].head())
```

```python
if "Availability" in df.columns:
    df["Availability"] = df["Availability"].str.replace("\n", " ", regex=True)

    df["Availability"] = df["Availability"].replace({
        "Unavailable We don't know when or if this item will be back in stock.": "Unavailable"
    })

    # Print the updated "Availability" column
    print("\nUpdated Availability Column:")
    print(df[["Availability"]].head())  # Display first few rows
```

```python
if "Availability" in df.columns:
    df["Availability"] = df["Availability"].str.strip().replace({
        "Only 1 left in stock.": "1 Left",
        "Only 2 left in stock.": "2 Left",
        "Currently unavailable.": "Unavailable",
        "We don’t know when or if this item will be back in stock.": "Unavailable",
        "In stock.": "In Stock",
        "Unknown": "Unavailable",
        "Unavailable": "Unavailable"  
    }, regex=True)

    # Print the updated "Availability" column
    print("\nUpdated Availability Column:")
    print(df[["Availability"]].head())  # Display first few rows
```
